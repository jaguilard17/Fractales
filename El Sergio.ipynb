```julia
using Fatou
using PyPlot
```

# Ejemplos
1. Funciones comunes:

- Polin√≥micas

```julia
c = -0.092 - 0.839im
nf = juliafill(:(z^2+$c),‚àÇ=[-1.5,1.5,-1.5,1.5],N=300,n=10000,cmap="cubehelix",iter=true)
plot(fatou(nf), bare=false)
basin(nf,1)
```

- Trigonom√©trica

```julia
nf = newton(:(tan(z)),m=1,‚àÇ=[-5,5,-5,5],n=500,N=300,iter=true,œµ=0.05,cmap="Greys")
nf |> fatou |> plot
basin(nf,2)
```

```julia
nf = newton(:(sin(z)),m=2im,‚àÇ=[-4,4,-4,4],n=800,N=500,iter=true,œµ=0.05,cmap="gist_earth")
nf |> fatou |> plot
basin(nf,2)
```

- Exponencial

```julia
nf = newton(:(2^(z)),m=-1,‚àÇ=[-100.,100.,-100.,100.],n=1000,N=3000,iter=true,œµ=0.06,cmap="cubehelix")
nf |> fatou |> plot
basin(nf,2)
```

- Logaritmo

```julia
nf = newton(:(log(z)),m=0.192-0.192im,‚àÇ=[-30.,30.,-30.,30.],n=500,N=43,iter=true,œµ=0.06,cmap="cubehelix")
nf |> fatou |> plot
basin(nf,2)
```
